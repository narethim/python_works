```python
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')
```

```python
slices = [60, 40]

plt.pie(slices)
plt.title('Basic Pie Chart')
plt.show()
```

# Add label

```python
slices = [60, 40]
labels = ['Category A', 'Category B']

# Add label
plt.pie(slices, labels=labels)
plt.title('Basic Pie Chart')

plt.show()
```

```python
slices = [60, 40]
labels = ['Category A', 'Category B']

# Add label and edgecolor
plt.pie(slices, labels=labels, wedgeprops={'edgecolor': 'black'})
plt.title('Basic Pie Chart')

plt.show()
```

```python
slices = [60, 40]
labels = ['Category A', 'Category B']
colors = ['#FF9999', '#6683FF']

# Add label, colors and edgecolor
plt.pie(slices, labels=labels, colors=colors, wedgeprops={'edgecolor': 'black'})
plt.title('Basic Pie Chart')

plt.show()
```

```python
slices = [60, 40]
labels = ['Category A', 'Category B']
colors = ['#FF9999', '#6683FF']

# Explode
explode = [0.1, 0]

# Add label, colors, explode, and edgecolor
plt.pie(slices, labels=labels, colors=colors, explode=explode, wedgeprops={'edgecolor': 'black'})
plt.title('Basic Pie Chart')

plt.show()
```

```python
slices = [60, 40]
labels = ['Category A', 'Category B']
colors = ['#FF9999', '#6683FF']
# Explode
explode = [0.1, 0]

# Add label, colors, shadow, explode, and edgecolor
plt.pie(slices, labels=labels, colors=colors, shadow=True, explode=explode, wedgeprops={'edgecolor': 'black'})
plt.title('Basic Pie Chart')

plt.show()
```

```python
# Change start angle
```

```python
slices = [60, 40]
labels = ['Category A', 'Category B']
colors = ['#FF9999', '#6683FF']
# Explode
explode = [0.1, 0]

# Add startangle=90 and autopct='%1.1f%%'
plt.pie(slices, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90, explode=explode, wedgeprops={'edgecolor': 'black'})
plt.title('Basic Pie Chart')

plt.show()
```
