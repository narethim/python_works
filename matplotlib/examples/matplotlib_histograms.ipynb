```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('fivethirtyeight')
```

```python
np.random.seed(36)

ages = np.random.randint(21, 63, size=20).tolist()
print(ages)
```

```python
np.random.seed(36)

ages = np.random.randint(21, 63, size=20).tolist()
print(ages)

plt.hist(ages)

plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.show()
```

```python
np.random.seed(36)

ages = np.random.randint(21, 63, size=20).tolist()
print(ages)

# Add edgecolor='black'
plt.hist(ages, edgecolor='black')

plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.show()
```

```python
np.random.seed(36)

ages = np.random.randint(21, 63, size=20).tolist()
print(ages)

# Add edgecolor='black'
counts, bin_edges, _ = plt.hist(ages, edgecolor='black')

print('Counts per bin: ', counts)
print('Bin edges     : ', bin_edges)

plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.show()
```

```python
np.random.seed(36)

ages = np.random.randint(21, 63, size=20).tolist()
print(ages)

# Add edgecolor='black'
counts, bin_edges, _ = plt.hist(ages, bins=5, edgecolor='black')

print('Counts per bin: ', counts)
print('Bin edges     : ', bin_edges)

plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.show()
```

```python
np.random.seed(36)

ages = np.random.randint(21, 63, size=20).tolist()
print(ages)

custom_bins = [20, 30, 40, 50, 60, 70]

# Add edgecolor='black'
counts, bin_edges, _ = plt.hist(ages, bins=custom_bins, edgecolor='black')

print('Counts per bin: ', counts)
print('Bin edges     : ', bin_edges)

plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.show()
```

```python
np.random.seed(36)

ages = np.random.randint(21, 63, size=20).tolist()
print(ages)

custom_bins = [20, 30, 40, 50, 60, 70]

median_age = np.median(ages)
# Add edgecolor='black'
counts, bin_edges, _ = plt.hist(ages, bins=custom_bins, edgecolor='black', log=True)
plt.axvline(median_age, color='red', linestyle='--', linewidth=1)
plt.text(median_age+1, 3, 'Median', color='red')

print('Counts per bin: ', counts)
print('Bin edges     : ', bin_edges)

plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.show()
```
