```python
import matplotlib.pyplot as plt
import pandas as pd

data = {
    'Date': pd.date_range(start='2025-01-01', periods=10, freq='D'),
    'Temperature': [30, 32, 35, 31, 30, 29, 33, 34, 36, 32],
    'Humidity': [70, 65, 60, 75, 80, 78, 72, 68, 64, 66]
}
df = pd.DataFrame(data)

df
```

```python
fig, ax = plt.subplots()

ax.plot(df['Date'], df['Temperature'], label='Temperature')
ax.set_xlabel('Date')
ax.set_ylabel('Temperature (C)')
ax.set_title('Daily Temperature')
ax.legend()

plt.show()
```

```python
fig, ax = plt.subplots(2, 1) # 2 rows, 1 column

ax[0].plot(df['Date'], df['Temperature'], label='Temperature')
ax[0].set_xlabel('Date')
ax[0].set_ylabel('Temperature (C)')
ax[0].set_title('Daily Temperature')
ax[0].legend()

ax[1].plot(df['Date'], df['Humidity'], label='Humidity', color='orange')
ax[1].set_xlabel('Date')
ax[1].set_ylabel('Humidity (%)')
ax[1].set_title('Daily Humidity')
ax[1].legend()

plt.show()
```

```python
fig, ax = plt.subplots(2, 1) # 2 rows, 1 column

ax[0].plot(df['Date'], df['Temperature'], label='Temperature')
ax[0].set_xlabel('Date')
ax[0].set_ylabel('Temperature (C)')
ax[0].set_title('Daily Temperature')
ax[0].legend()
ax[0].tick_params(axis='x', rotation=45) # 45 degree

ax[1].plot(df['Date'], df['Humidity'], label='Humidity', color='orange')
ax[1].set_xlabel('Date')
ax[1].set_ylabel('Humidity (%)')
ax[1].set_title('Daily Humidity')
ax[1].legend()
ax[1].tick_params(axis='x', rotation=45) # 45 degree

plt.show()
```

```python
fig, ax = plt.subplots(2, 1) # 2 rows, 1 column

ax[0].plot(df['Date'], df['Temperature'], label='Temperature')
ax[0].set_xlabel('Date')
ax[0].set_ylabel('Temperature (C)')
ax[0].set_title('Daily Temperature')
ax[0].legend()
ax[0].tick_params(axis='x', rotation=45) # 45 degree

ax[1].plot(df['Date'], df['Humidity'], label='Humidity', color='orange')
ax[1].set_xlabel('Date')
ax[1].set_ylabel('Humidity (%)')
ax[1].set_title('Daily Humidity')
ax[1].legend()
ax[1].tick_params(axis='x', rotation=45) # 45 degree

plt.tight_layout() # tight_layout
plt.show()
```

# Share axis

```python
fig, ax = plt.subplots(2, 1, sharex=True) # 2 rows, 1 column, share x axis

ax[0].plot(df['Date'], df['Temperature'], label='Temperature')
ax[0].set_ylabel('Temperature (C)')
ax[0].legend()

ax[1].plot(df['Date'], df['Humidity'], label='Humidity', color='orange')
ax[1].set_xlabel('Date')
ax[1].set_ylabel('Humidity (%)')
ax[1].legend()
ax[1].tick_params(axis='x', rotation=45) # 45 degree

fig.suptitle('Temperature and Humidity Over Time') # Figure title

plt.tight_layout() # tight_layout
plt.show()
```

# Another way: 2 separate figures

```python
fig1 = plt.figure()
ax1 = fig1.add_subplot(111)

ax1.plot(df['Date'], df['Temperature'], label='Temperature')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (C)')
ax1.set_title('Temperature Over Time')
ax1.legend()
ax1.tick_params(axis='x', rotation=45)

fig2 = plt.figure()
ax2 = fig2.add_subplot(111)

ax2.plot(df['Date'], df['Humidity'], label='Humidity', color='orange')
ax2.set_xlabel('Date')
ax2.set_ylabel('Humidity (%)')
ax1.set_title('Humidity Over Time')
ax2.legend()
ax2.tick_params(axis='x', rotation=45) # 45 degree

plt.tight_layout() # tight_layout
plt.show()
```
